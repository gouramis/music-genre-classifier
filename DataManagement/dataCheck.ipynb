```python
def makeKfold(df, k):
    kf = skl.model_selection.KFold(n_splits = k, shuffle = True)
    df_collection = {}
    i = 0
    for train_index, test_index in kf.split(df):
        df_collection[i,"training"] = df.iloc[train_index]
        df_collection[i,"testing"] = df.iloc[test_index]
        i = i + 1
    return df_collection
```
